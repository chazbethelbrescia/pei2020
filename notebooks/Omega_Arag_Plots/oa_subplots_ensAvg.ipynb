# Plots for Annual Mean and Minimum Omega Aragonite in 1990 and 2100 (Using RCP85 Ensemble Average)

## Annual Mean Omega Aragonite

```python
import xarray as xr
from matplotlib import pyplot as plt
from matplotlib import gridspec
from cartopy import crs as ccrs
import cartopy.feature as cfeature
from cartopy.util import add_cyclic_point
```

```python
%reload_ext autoreload
%autoreload 2
from chazbpei2020.preprocessing import *
```

## Surface irr0.1 (1990 vs. 2100)

```python
rootdir = '~/chazbpei2020/data/processed/Omega_Arag'
experiment = 'omega_arag_irr_0.1Wm2_ensAvg_1950_2100.nc'

path = rootdir+'/'+experiment

# Obtain the ensemble average (preprocessed)
ds_ensembleAvg = xr.open_dataset(path)
ds_ensembleAvg
# Specify variables
X = ds_ensembleAvg['XT_OCEAN']
Y = ds_ensembleAvg['YT_OCEAN']
Z0 = annual_mean(ds_ensembleAvg, 'OMEGA_ARAG_IRR', 1990) 
Z1 = annual_mean(ds_ensembleAvg, 'OMEGA_ARAG_IRR', 2100)

# call contour_default function specifying variables
fig = contour_subplotsh(X, Y, Z0, Z1, 12, 4, central_longitude=180)
fig.suptitle("Mean Annual Omega Aragonite at Surface irr0.1", va='top', fontsize=22)

ax0 = fig.get_axes()[0]
ax1 = fig.get_axes()[1]
ax0.set_title('1990', fontsize=20)
ax1.set_title('2100', fontsize=20)

# Define contour levels for Omega Aragonite
clevs=[0,1,2,3,4]
# Redefine projection for offset (offset by 180)
crs_offset = ccrs.PlateCarree(central_longitude=0)

im0 = ax0.contourf(X,Y,Z0, levels=clevs, transform=crs_offset)
im1 = ax1.contourf(X,Y,Z1, levels=clevs, transform=crs_offset)

cbar = plt.colorbar(im0, ax=fig.get_axes(), orientation='horizontal', fraction=0.05, pad=0.05)
cbar.set_label('Omega Aragonite',fontsize=12)

fig.savefig("omega_arag_subplots")
```
