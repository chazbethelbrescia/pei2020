# Plots for Annual Mean and Minimum Omega Aragonite in 1990 and 2100 (Using RCP85 Ensemble Average and Minimum)

```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import gridspec
from cartopy import crs as ccrs
import cartopy.feature as cfeature
from cartopy.util import add_cyclic_point
```

# Surface irr0.1

```python
# irr0.1 ensemble average (preprocessed)
directory = '~/chazbpei2020/data/processed/Omega_Arag'
filename = 'omega_arag_irr_0.1Wm2_ensAvg_1950_2100.nc'
irrAvg_path = directory+'/'+filename
ds_irrAvg = xr.open_dataset(irrAvg_path).rename({'OMEGA_ARAG_IRR': 'OMEGA_ARAG'})

# Get Datasets for annual mean and min
ds_irrAvgMean = ds_irrAvg.groupby('TIME.year').mean(dim='TIME')
ds_irrAvgMin = ds_irrAvg.groupby('TIME.year').min(dim='TIME')
```

```python
# irr0.1 ensemble minimum (preprocessed)
directory = '~/chazbpei2020/data/processed/Omega_Arag'
filename = 'omega_arag_irr_0.1Wm2_ensMin_1950_2100.nc'
irrMin_path = directory+'/'+filename
ds_irrMin = xr.open_dataset(irrMin_path).rename({'OMEGA_ARAG_IRR': 'OMEGA_ARAG'})

# Get Datasets for annual mean and min
ds_irrMinMean = ds_irrMin.groupby('TIME.year').mean(dim='TIME')
ds_irrMinMin = ds_irrMin.groupby('TIME.year').min(dim='TIME')
```

## Annual Mean Omega Aragonite - Ensemble Average (1990 vs. 2100)

```python
# maps of irr0.1 ensemble average
crs = ccrs.Robinson(central_longitude=180) # Specify projection
fig, axs = plt.subplots(figsize=(18,10),nrows=2,ncols=2, subplot_kw={'projection':crs})

fig.suptitle('RCP8.5 Ensemble Average - $\Omega$ Aragonite, Surface irr0.1W/m$^2$', 
             fontsize=22)

# Subplots for both annual min and mean
contour_oa(ds_irrAvgMean, 1990, axs[0,0], 'Annual Mean')
contour_oa(ds_irrAvgMean, 2100, axs[0,1], 'Annual Mean')
contour_oa(ds_irrAvgMin, 1990, axs[1,0], 'Annual Minimum')
contour_oa(ds_irrAvgMin, 2100, axs[1,1], 'Annual Minimum')

fig.savefig("omega_arag_plots")
```

## 2000 vs. 2100

```python
# maps of irr0.1 ensemble average
crs = ccrs.Robinson(central_longitude=180) # Specify projection
fig, axs = plt.subplots(figsize=(18,5.5),nrows=1,ncols=2, subplot_kw={'projection':crs})

fig.suptitle('RCP8.5 Ensemble Average - $\Omega$ Aragonite at 0.1W/m$^2$ Irradiance Depth', 
             fontsize=22)

# Subplots for both annual min and mean
contour_oa(ds_irrAvgMean, 2000, axs[0], 'Annual Mean')
contour_oa(ds_irrAvgMean, 2100, axs[1], 'Annual Mean')

fig.savefig("omega_arag_plots")
```

# Historical Omega Aragonite (1950-2019)

## Surface irr0.1

```python
# Calculate the time-mean temperature for 1950-2019
ds_oa_mean = ds_irrAvgMean.sel(year=1950).copy()
for yr in range(1951,2020):
     ds_oa_mean += ds_irrAvgMean.sel(year=yr)
        
ds_oa_mean = ds_oa_mean/70
```

```python
# Define projection, coordinates, levels for plots
crs = ccrs.Robinson(central_longitude=180)

lon = ds_oa_mean.XT_OCEAN.data
lat = ds_oa_mean.YT_OCEAN.data

levels = [0,1,2,3,4]

def contour_omega():
    im = ax.contourf(lon,lat, ds_oa_mean.OMEGA_ARAG.squeeze(), levels=levels, 
                     transform=ccrs.PlateCarree(), robust=True)
    cbar = plt.colorbar(im,ax=ax,orientation='horizontal',fraction=0.05,pad=0.05)
    cbar.set_label('$\Omega$ Aragonite', fontsize=16)
    ax.add_feature(cfeature.LAND,zorder=10,facecolor='darkgray')
```

```python
# Maps irr0.1 Ensemble Average
fig, ax = plt.subplots(figsize=(18,10), subplot_kw={'projection':crs})

fig.suptitle('RCP8.5 Omega Aragonite, Ensemble Average, Surface irr0.1', 
             fontsize=22)

# Contour plot and colorbar
contour_omega()

fig.savefig("omega_arag_plots")
```
