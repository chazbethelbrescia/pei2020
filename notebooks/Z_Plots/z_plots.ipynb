## Habitable Depths (Using Irradiance and Omega Arag Thresholds)

```python
crs = ccrs.Robinson(central_longitude=180)
fig, ax = plt.subplots(figsize=[16,8],
                               subplot_kw={'projection':crs})

# create new dataset with overlayed depth thresholds
ds_irrZ = ds_irrZAvgMean.sel(year=1950)
ds_omegaZ = ds_omegaZAvgMean.sel(year=1950)


lon = ds_irrZ.XT_OCEAN.data
lat = ds_irrZ.YT_OCEAN.data
nx = len(lon)
ny = len(lat)

da_habitable = np.zeros(shape=(nx,ny))
for x in range(nx):
    for y in range(ny):
        # tally number of factors that satisfy habitable conditions
        irr = ds_irrZ.sel(XT_OCEAN=lon[x], YT_OCEAN=lat[y]).Z.data
        omega_arag = ds_omegaZ.sel(XT_OCEAN=lon[x], YT_OCEAN=lat[y]).Z.data
        da_habitable[x,y] = min(irr, omega_arag)  # take the minimum depth

# Create DataArray and plot habitable depths for the decade
da_habitable = xr.DataArray(da_habitable, dims=['xt_ocean','yt_ocean'], coords=[lon,lat])

da_habitable.T.plot(ax=ax, transform=ccrs.PlateCarree(), robust=False)
ax.set_title('Habitable Depths in 1950s (m)',fontsize=20)
ax.set_global()

fig.savefig('./z_plots/z_habitable_1950s')
```
