# Plots for Annual Mean and Minimum Temperature in 1990 and 2100 (Using RCP85 Ensemble)

```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import gridspec
from cartopy import crs as ccrs
import cartopy.feature as cfeature
from cartopy.util import add_cyclic_point
```

```python
%matplotlib inline
```

```python
%reload_ext autoreload
%autoreload 2
from chazbpei2020.preprocessing import *
```

# Surface irr0.1

```python
# irr0.1 ensemble average (preprocessed)
directory = '~/chazbpei2020/data/processed/Temp'
filename = 'temp_irr_0.1Wm2_ensAvg_1950_2100.nc'
irrAvg_path = directory+'/'+filename
ds_irrAvg = xr.open_dataset(irrAvg_path).rename({'XT_OCEAN': 'xt_ocean',
                                     'YT_OCEAN': 'yt_ocean',
                                     'TIME': 'time',
                                     'TEMP_IRR': 'temp'})

# Get Datasets for annual mean and min
ds_irrAvgMean = ds_irrAvg.groupby('time.year').mean(dim='time')
ds_irrAvgMin = ds_irrAvg.groupby('time.year').min(dim='time')
```

```python
# irr0.1 ensemble minimum (preprocessed)
directory = '~/chazbpei2020/data/processed/Temp'
filename = 'temp_irr_0.1Wm2_ensMin_1950_2100.nc'
irrMin_path = directory+'/'+filename
ds_irrMin = xr.open_dataset(irrMin_path).rename({'XT_OCEAN': 'xt_ocean',
                                     'YT_OCEAN': 'yt_ocean',
                                     'TIME': 'time',
                                     'TEMP_IRR': 'temp'})

# Get Datasets for annual mean and min
ds_irrMinMean = ds_irrMin.groupby('time.year').mean(dim='time')
ds_irrMinMin = ds_irrMin.groupby('time.year').min(dim='time')
```

## Annual Mean Temperature - Ensemble Average (1990 vs. 2100)

```python
# maps of irr0.1 ensemble average
crs = ccrs.Robinson(central_longitude=180) # Specify projection
fig, axs = plt.subplots(figsize=(18,10),nrows=2,ncols=2, subplot_kw={'projection':crs})
# Specify a figure title
fig.suptitle('RCP8.5 Temperature, Ensemble Average, Surface irr0.1', 
             fontsize=22)

# Subplots for both annual min and mean
contour_temp(ds_irrAvgMean, 1990, axs[0,0], 'Annual Mean')
contour_temp(ds_irrAvgMean, 2100, axs[0,1], 'Annual Mean')
contour_temp(ds_irrAvgMin, 1990, axs[1,0], 'Annual Minimum')
contour_temp(ds_irrAvgMin, 2100, axs[1,1], 'Annual Minimum')

fig.savefig("temp_plots")
```

## 2000 vs. 2100

```python
# maps of irr0.1 ensemble average
crs = ccrs.Robinson(central_longitude=180) # Specify projection
fig, axs = plt.subplots(figsize=(18,5),nrows=1,ncols=2, subplot_kw={'projection':crs})

fig.suptitle('RCP8.5 Ensemble Average - Temperature at 0.1W/m$^2$ Irradiance Depth', 
             fontsize=22)

# Subplots for both annual min and mean
contour_temp(ds_irrAvgMean, 2000, axs[0], 'Annual Mean')
contour_temp(ds_irrAvgMean, 2100, axs[1], 'Annual Mean')


fig.savefig("temp_plots")
```

## Annual Mean Omega Aragonite - Ensemble Minimum (1990 vs. 2100)

```python
# maps of irr0.1 ensemble minimum
crs = ccrs.Robinson(central_longitude=180) # Specify projection
fig, axs = plt.subplots(figsize=(18,10),nrows=2,ncols=2, subplot_kw={'projection':crs})
# Specify figure title
fig.suptitle('RCP8.5 Temperature, Ensemble Minimum, Surface irr0.1', 
             fontsize=22)

contour_temp(ds_irrMinMean, 1990, axs[0,0], 'Annual Mean')
contour_temp(ds_irrMinMean, 2100, axs[0,1], 'Annual Mean')
contour_temp(ds_irrMinMin, 1990, axs[1,0], 'Annual Minimum')
contour_temp(ds_irrMinMin, 2100, axs[1,1], 'Annual Minimum')

fig.savefig("temp_plots")
```
